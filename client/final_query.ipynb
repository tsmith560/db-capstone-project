```python
query = """
SELECT 
    CustomerDetails.Name AS CustomerName,
    CustomerDetails.ContactNumber AS ContactNumber,
    Orders.TotalPrice AS OrderTotal
FROM 
    CustomerDetails
INNER JOIN 
    Orders 
ON 
    CustomerDetails.CustomerID = Orders.CustomerID
WHERE 
    Orders.TotalPrice > 60;
"""
```

```python
cursor.execute(query)
results = cursor.fetchall()

for row in results:
    print(row)
```
